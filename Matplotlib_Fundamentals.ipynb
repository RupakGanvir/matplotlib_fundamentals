## **Importing Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## **Basics**

```python
# The most basic plot

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

```python
# plt.plot() just draws a line by default
# Let's add a title and labels to make it useful

plt.plot(x, y)
plt.title('My First Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

## **Figure and Axes**

There are two ways to create plots in matplotlib:
1) Using plt directly (like above) - quick and simple
2) Using fig and ax objects - more control, recommended way

```python
# Way 1: Using plt directly (simple, quick)

plt.plot(x, y)
plt.title('Simple Way')
plt.show()
```

```python
# Way 2: Using fig and ax (the proper way)
# fig is the whole window/canvas
# ax is the actual plot inside the canvas

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title('Proper Way')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

plt.show()
```

```python
# figsize controls the size of the figure (width, height) in inches

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y)
ax.set_title('Custom Size')

plt.show()
```

## **Line Plot**

```python
x = np.linspace(0, 10, 100) # 100 evenly spaced values from 0 to 10
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y)
ax.set_title('Sin Wave')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
plt.show()
```

```python
# We can plot multiple lines on the same axes

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, np.sin(x)) # First line
ax.plot(x, np.cos(x)) # Second line on same axes

ax.set_title('Sin and Cos')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
# Adding colors, labels, and legend

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, np.sin(x), color='blue', label='sin(x)')
ax.plot(x, np.cos(x), color='red', label='cos(x)')

ax.set_title('Sin and Cos')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend() # shows the labels
plt.show()
```

### **Line Styles and Markers**

```python
# linestyle controls the style of the line
# '-' is solid (default), '--' is dashed, ':' is dotted, '-.' is dash-dot

x = np.linspace(0, 5, 50)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, x,    linestyle='-',  label='solid')
ax.plot(x, x+1,  linestyle='--', label='dashed')
ax.plot(x, x+2,  linestyle=':',  label='dotted')
ax.plot(x, x+3,  linestyle='-.', label='dash-dot')

ax.legend()
ax.set_title('Line Styles')
plt.show()
```

```python
# marker adds a symbol at each data point

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y, marker='o', color='blue', label='circle')   # o = circle
ax.plot(x, [i+2 for i in y], marker='s', color='red', label='square')   # s = square
ax.plot(x, [i+4 for i in y], marker='^', color='green', label='triangle') # ^ = triangle

ax.legend()
ax.set_title('Markers')
plt.show()
```

```python
# linewidth controls the thickness of the line
# alpha controls the transparency (0 = invisible, 1 = solid)

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, np.sin(x), linewidth=1, label='lw=1')
ax.plot(x, np.sin(x)+1, linewidth=3, label='lw=3')
ax.plot(x, np.sin(x)+2, linewidth=3, alpha=0.3, label='lw=3, alpha=0.3') # transparent

ax.legend()
ax.set_title('Linewidth and Alpha')
plt.show()
```

## **Scatter Plot**

```python
# Scatter plot is used to show the relationship between two variables
# Instead of connecting points with a line, it just shows the points

x = np.random.randn(100) # 100 random values from normal distribution
y = np.random.randn(100)

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(x, y)
ax.set_title('Basic Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
# s controls the size of each point
# c controls the color
# alpha controls transparency

x = np.random.randn(200)
y = np.random.randn(200)

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(x, y, s=50, c='red', alpha=0.5) # s=50 makes points bigger, alpha=0.5 makes them semi-transparent
ax.set_title('Scatter with custom size and color')
plt.show()
```

```python
# We can also color each point differently based on a third variable
# c takes an array and cmap (colormap) decides how colors are assigned

x = np.random.randn(200)
y = np.random.randn(200)
colors = np.random.rand(200) # a random value for each point that decides its color

fig, ax = plt.subplots(figsize=(7, 5))

scatter = ax.scatter(x, y, c=colors, cmap='viridis', s=50, alpha=0.7) # viridis is a built-in colormap

plt.colorbar(scatter, ax=ax) # shows the colorbar on the side to tell what each color means
ax.set_title('Scatter with Colormap')
plt.show()
```

## **Bar Plot**

```python
# Bar plot compares values across categories

categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 12, 67, 34]

fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(categories, values)
ax.set_title('Basic Bar Plot')
ax.set_xlabel('Category')
ax.set_ylabel('Value')
plt.show()
```

```python
# We can change bar colors

fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(categories, values, color=['blue', 'red', 'green', 'orange', 'purple'])
ax.set_title('Colored Bars')
plt.show()
```

```python
# Horizontal bar chart using barh()

fig, ax = plt.subplots(figsize=(7, 4))
ax.barh(categories, values, color='steelblue') # barh = horizontal bar
ax.set_title('Horizontal Bar Plot')
ax.set_xlabel('Value')
ax.set_ylabel('Category')
plt.show()
```

```python
# Grouped bar chart - comparing two groups side by side

departments = ['IT', 'HR', 'Sales', 'Legal']
salary_2023 = [70000, 50000, 60000, 90000]
salary_2024 = [75000, 53000, 65000, 95000]

x = np.arange(len(departments)) # [0, 1, 2, 3]
width = 0.35 # width of each bar

fig, ax = plt.subplots(figsize=(8, 4))

ax.bar(x - width/2, salary_2023, width, label='2023') # shift left by half width
ax.bar(x + width/2, salary_2024, width, label='2024') # shift right by half width

ax.set_xticks(x) # place ticks at 0, 1, 2, 3
ax.set_xticklabels(departments) # label them with department names
ax.set_title('Salary Comparison 2023 vs 2024')
ax.set_ylabel('Salary')
ax.legend()

plt.show()
```

## **Histogram**

```python
# Histogram shows the distribution of a single variable
# It groups values into bins and counts how many fall in each bin

data = np.random.randn(1000) # 1000 random values

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(data) # by default it creates 10 bins
ax.set_title('Basic Histogram')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
plt.show()
```

```python
# bins controls how many groups the data is divided into

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(data, bins=30, color='steelblue', edgecolor='black') # edgecolor adds a border to each bar
ax.set_title('Histogram with 30 bins')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
plt.show()
```

```python
# Overlapping histograms to compare two distributions

data1 = np.random.normal(0, 1, 500)  # mean=0, std=1
data2 = np.random.normal(2, 1, 500)  # mean=2, std=1 (shifted to the right)

fig, ax = plt.subplots(figsize=(8, 4))

ax.hist(data1, bins=30, alpha=0.5, label='Group 1') # alpha=0.5 so we can see through them
ax.hist(data2, bins=30, alpha=0.5, label='Group 2')

ax.set_title('Overlapping Histograms')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()
plt.show()
```

## **Subplots**

subplots() lets us create multiple plots in the same figure

```python
# 1 row, 2 columns → two plots side by side

fig, axes = plt.subplots(1, 2, figsize=(12, 4)) # axes is now an array of 2 axes objects

x = np.linspace(0, 10, 100)

axes[0].plot(x, np.sin(x), color='blue')
axes[0].set_title('Sin')

axes[1].plot(x, np.cos(x), color='red')
axes[1].set_title('Cos')

plt.show()
```

```python
# 2 rows, 2 columns → four plots in a grid
# axes is now a 2D array, so we access with axes[row, col]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Sin')

axes[0, 1].plot(x, np.cos(x), color='red')
axes[0, 1].set_title('Cos')

axes[1, 0].plot(x, np.tan(x), color='green')
axes[1, 0].set_title('Tan')
axes[1, 0].set_ylim(-5, 5) # tan goes to infinity so we limit the y axis

axes[1, 1].plot(x, np.exp(-x), color='orange')
axes[1, 1].set_title('e^(-x)')

plt.tight_layout() # automatically adjusts spacing so titles don't overlap
plt.show()
```

## **Pie Chart**

```python
labels  = ['Python', 'Java', 'C++', 'JavaScript', 'Others']
sizes   = [35, 25, 20, 15, 5]

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(sizes, labels=labels)
ax.set_title('Programming Language Popularity')
plt.show()
```

```python
# autopct shows the percentage on each slice
# explode separates a slice from the rest

explode = (0.1, 0, 0, 0, 0) # only the first slice (Python) is separated

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, startangle=90)
ax.set_title('Programming Language Popularity')
plt.show()
```

## **Heatmap**

```python
# A heatmap shows data as colors in a 2D grid
# imshow() is used to create heatmaps in matplotlib

data = np.random.rand(5, 5) # 5x5 grid of random values

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(data, cmap='hot') # cmap decides the color scheme
plt.colorbar(im, ax=ax) # shows the scale
ax.set_title('Basic Heatmap')
plt.show()
```

```python
# Common use: Confusion Matrix

confusion_matrix = np.array([[50, 3,  2],
                              [5,  45, 5],
                              [1,  4,  48]])

fig, ax = plt.subplots(figsize=(6, 5))

im = ax.imshow(confusion_matrix, cmap='Blues')
plt.colorbar(im, ax=ax)

# Add the numbers inside each cell
for i in range(3):
    for j in range(3):
        ax.text(j, i, str(confusion_matrix[i, j]), ha='center', va='center', fontsize=12)

ax.set_xticks([0, 1, 2])
ax.set_yticks([0, 1, 2])
ax.set_xticklabels(['Cat', 'Dog', 'Bird'])
ax.set_yticklabels(['Cat', 'Dog', 'Bird'])
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix')
plt.show()
```

```python
# Another common use: Correlation Heatmap

# Create some correlated data
df = pd.DataFrame({
    'Age':    [25, 30, 35, 40, 45, 50, 55],
    'Income': [30000, 45000, 55000, 70000, 80000, 90000, 95000],
    'Score':  [60, 65, 70, 75, 72, 68, 65],
    'Debt':   [5000, 8000, 12000, 15000, 10000, 7000, 4000]
})

corr = df.corr() # calculate the correlation between all columns
print(corr)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))

im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1) # vmin and vmax fix the scale from -1 to 1
plt.colorbar(im, ax=ax)

# Add correlation values inside each cell
for i in range(len(corr)):
    for j in range(len(corr.columns)):
        ax.text(j, i, f'{corr.iloc[i,j]:.2f}', ha='center', va='center', fontsize=10)

ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45)
ax.set_yticklabels(corr.columns)
ax.set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

## **Customizing Plots**

```python
# Axis limits - controlling what range is shown

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x))

ax.set_xlim(0, 5)   # show only x from 0 to 5
ax.set_ylim(-2, 2)  # show y from -2 to 2

ax.set_title('Custom Axis Limits')
plt.show()
```

```python
# Custom ticks - controlling where the tick marks are and what they say

x = np.linspace(0, 2*np.pi, 100)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x))

ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]) # where to put ticks
ax.set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])     # what to label them

ax.set_title('Custom Tick Labels')
plt.show()
```

```python
# Adding grid lines

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x))
ax.grid(True) # turns on the grid
ax.set_title('Plot with Grid')
plt.show()
```

```python
# Adding annotations - pointing to a specific part of the plot

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y)

# annotate() adds text with an arrow pointing to a coordinate
ax.annotate('Maximum', 
            xy=(np.pi/2, 1),           # point to annotate (tip of the arrow)
            xytext=(np.pi/2 + 1, 0.7), # where the text goes (tail of the arrow)
            arrowprops=dict(arrowstyle='->'))

ax.set_title('Plot with Annotation')
plt.show()
```

## **Saving a Figure**

Always call savefig() before show(). If you call show() first, the figure gets cleared and savefig() will save a blank image.

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x), color='blue', label='sin(x)')
ax.set_title('Sin Wave')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.legend()

plt.tight_layout()
fig.savefig('sin_plot.png', dpi=150, bbox_inches='tight') # dpi=150 gives good quality, bbox_inches='tight' prevents labels from being cut off
plt.show()

print('Plot saved!')
```

## **Plotting Directly from Pandas**

```python
# Pandas has a built in .plot() method that uses matplotlib under the hood
# It's a quick way to plot without much code

df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [100, 120, 95, 140, 160, 130],
    'Returns': [10, 15, 8, 20, 18, 12]
})

df.set_index('Month', inplace=True)
df
```

```python
# Line plot from DataFrame

df.plot(figsize=(8, 4), title='Monthly Sales and Returns')
plt.ylabel('Units')
plt.show()
```

```python
# Bar plot from DataFrame

df.plot(kind='bar', figsize=(8, 4), title='Monthly Comparison')
plt.ylabel('Units')
plt.xticks(rotation=0) # keeps x labels horizontal
plt.show()
```

```python
# We can also plot just one column

df['Sales'].plot(kind='line', figsize=(8, 4), title='Sales Only', marker='o')
plt.ylabel('Sales')
plt.show()
```

## **Exercise**
1) Create a 2x2 subplot figure with:
   - Top Left: Line plot of y = x^2 from x = -5 to 5
   - Top Right: Scatter plot of 100 random points
   - Bottom Left: Bar plot of 5 categories with random values
   - Bottom Right: Histogram of 500 random normal values

2) Plot sin(x) and sin(2x) on the same axes from 0 to 2π. Add a title, labels, legend, and grid. Color sin(x) blue and sin(2x) red with a dashed line.

```python
# Solution 1:

np.random.seed(42)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Top Left - Line plot of y = x^2
x = np.linspace(-5, 5, 100)
axes[0, 0].plot(x, x**2, color='blue')
axes[0, 0].set_title('y = x²')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')

# Top Right - Scatter plot
axes[0, 1].scatter(np.random.randn(100), np.random.randn(100), alpha=0.6)
axes[0, 1].set_title('Random Scatter')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')

# Bottom Left - Bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, 5)
axes[1, 0].bar(categories, values, color='steelblue')
axes[1, 0].set_title('Category Values')
axes[1, 0].set_xlabel('Category')
axes[1, 0].set_ylabel('Value')

# Bottom Right - Histogram
axes[1, 1].hist(np.random.randn(500), bins=25, color='green', edgecolor='black')
axes[1, 1].set_title('Normal Distribution')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Solution 2:

x = np.linspace(0, 2*np.pi, 300)

fig, ax = plt.subplots(figsize=(9, 4))

ax.plot(x, np.sin(x), color='blue', label='sin(x)')
ax.plot(x, np.sin(2*x), color='red', linestyle='--', label='sin(2x)')

ax.set_title('sin(x) vs sin(2x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```

## Key Takeaways

- `plt.plot()` for quick plots, `fig, ax = plt.subplots()` for proper control
- `figsize=(width, height)` controls the canvas size in inches
- Line plot → trends, Scatter → relationships, Bar → comparisons, Histogram → distributions, Pie → proportions
- `ax.legend()` only works if you set `label=` in the plot
- `plt.tight_layout()` prevents titles and labels from overlapping
- Always call `fig.savefig()` before `plt.show()` or you'll save a blank image
- `plt.subplots(rows, cols)` creates a grid of plots, access with `axes[row, col]`
- `imshow()` is used for heatmaps, colorbar shows what the colors mean
- Pandas `.plot()` is just a shortcut that calls matplotlib under the hood
